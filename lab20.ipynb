```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris

# Load the dataset (replace this with your own dataset)
data = load_iris()
X = data.data  # Features
y = data.target  # Target labels (if you want to color by label)

# Step 1: Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 2: Perform PCA
pca = PCA(n_components=2)  # We want the first two principal components
X_pca = pca.fit_transform(X_scaled)

# Step 3: Plot the first two principal components
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=50)
plt.colorbar(label='Target class')  # This adds a color bar to indicate class labels
plt.title('PCA - First Two Principal Components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()

# Step 4: Print explained variance ratio
print(f'Explained variance ratio for the first two components: {pca.explained_variance_ratio_}')
```
