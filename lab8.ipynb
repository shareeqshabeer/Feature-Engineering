```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create a sample dataset with a skewed 'Distance' column
data = {'Distance': [2, 3, 7, 15, 35, 40, 80, 150, 300, 600, 1000, 5000]}
df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)
```

```python
# Plot the distribution of the original 'Distance' column
sns.histplot(df['Distance'], kde=True)
plt.title('Original Distribution of Distance')
plt.xlabel('Distance')
plt.ylabel('Frequency')
plt.show()

# Calculate and print the skewness
skewness = df['Distance'].skew()
print(f"Skewness of 'Distance' before transformation: {skewness}")
```

```python
# Apply the natural logarithm transformation to 'Distance'
df['Log_Distance'] = np.log(df['Distance'])

# Display the DataFrame with the transformed data
print("\nDataFrame with Log-transformed 'Distance':")
print(df)
```

```python
# Plot the distribution of the transformed 'Log_Distance' column
sns.histplot(df['Log_Distance'], kde=True)
plt.title('Log-transformed Distribution of Distance')
plt.xlabel('Log(Distance)')
plt.ylabel('Frequency')
plt.show()

# Calculate and print the skewness after transformation
skewness_after = df['Log_Distance'].skew()
print(f"Skewness of 'Log_Distance' after transformation: {skewness_after}")
```
